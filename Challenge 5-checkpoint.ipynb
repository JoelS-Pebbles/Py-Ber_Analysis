```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = r"C:\Users\12109\Class\Py-Ber_Analysis\city_data.csv"
ride_data_to_load = r"C:\Users\12109\Class\Py-Ber_Analysis\ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

```python
#  1. Get the total rides for each city type
type_ride_count=pyber_data_df.groupby(["type"]).count()["ride_id"]
type_ride_count
```

```python
urban_data_df=pyber_data_df[pyber_data_df["type"]=="Urban"]
urban_data_df.head()
```

```python
suburban_data_df=pyber_data_df[pyber_data_df["type"]=="Suburban"]
suburban_data_df.head()
```

```python
rural_data_df=pyber_data_df[pyber_data_df["type"]=="Rural"]
rural_data_df.head()
```

```python
# 2. Get the total drivers for each city type
driver_ride_count = city_data_df.groupby(["type"]).sum()["driver_count"]
driver_ride_count
```

```python
#  3. Get the total amount of fares for each city type
sum_fare_by_type=pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fare_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
urban_avg_fare=urban_data_df.groupby(["type"]).mean()["fare"]
urban_avg_fare
```

```python
suburban_avg_fare=suburban_data_df.groupby(["type"]).mean()["fare"]
suburban_avg_fare
```

```python
rural_avg_fare=rural_data_df.groupby(["type"]).mean()["fare"]
rural_avg_fare
```

```python
avg_fare_per_ride=pyber_data_df.groupby(["type"]).mean()["fare"]
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver_by_city=sum_fare_by_type/driver_ride_count
avg_fare_per_driver_by_city
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df=pd.DataFrame({
      "Total Rides":type_ride_count,
      "Total Drivers": driver_ride_count,
      "Total Fares":sum_fare_by_type,
      "Average Fare per Ride":avg_fare_per_ride,
      "Average Fare per Driver":avg_fare_per_driver_by_city})
pyber_summary_df
```

```python
pyber_summary_df.index.name = None
```

```python
pyber_summary_df
```

```python
pyber_summary_df.dtypes
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"]=pyber_summary_df["Total Fares"].astype(float)

pyber_summary_df["Total Fares"]=pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"]=pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"]=pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df["Total Rides"]=pyber_summary_df["Total Rides"].map("{:,.0f}".format)
pyber_summary_df["Total Drivers"]=pyber_summary_df["Total Drivers"].map("{:,.0f}".format)
pyber_summary_df
```
